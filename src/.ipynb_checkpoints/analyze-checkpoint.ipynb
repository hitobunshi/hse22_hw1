```python
import re

def analyze(file):
    lengths = []
    cur_length = 0
    max_length = 0

    cur_seq = ''
    max_seq = ''

    num_seqs = 0

    for line in file:
        if line[0] == '>':
            num_seqs += 1

            if num_seqs != 1:
                lengths.append(cur_length)

            if cur_length > max_length:
                max_length = cur_length
                max_seq = cur_seq

            cur_seq = ''
            length = 0

        else:
            cur_seq += line.strip()
            cur_length += len(line.strip())

    if cur_length > 0:
        lengths.append(cur_length)
        if cur_length > max_length:
            max_length = cur_length
            max_seq = cur_seq

    total_length = sum(lengths)

    n50_score = 0
    n50 = 0
    lengths.sort(reverse=True)
    for length in lengths:
        n50_score += length
        if n50_score >= total_length / 2:
            n50 = length
            break
            
    return num_seqs, total_length, n50, max_length, max_seq


def print_info(num_seqs, total_length, n50, max_length, max_seq):
    print('Общее кол-во:', num_seqs)
    print('Общая длина:', total_length)
    print('Наибольшая длина:', max_length)
    print('N50:', n50)
    

def print_gap_info(num_seqs, total_length, n50, max_length, max_seq):
    print('Общая длина гэпов:', max_seq.count('N'))
    re.sub(r'N{2,}', 'N', max_seq)
    print('Количество гэпов:', max_seq.count('N'))
```

```python
print('Анализ контигов')
with open('../data/contigs.fasta') as file:
    res = analyze(file)
    print_info(*res)
    
print('\n\n')

with open('../data/scaffolds_before_gap_closed.fasta') as file:
    res = analyze(file)

    print('Анализ скаффолдов')
    print_info(*res)

    print('\nАнализ самого длинного скаффолда')
    print_gap_info(*res)

print('\n\n')

print('Анализ самого длинного скаффолда после gap_closed')
with open('../data/scaffolds.fasta') as file:
    res = analyze(file)
    print_gap_info(*res)
    
    with open('../data/longest.fasta', 'w') as output_file:
        output_file.write(res[-1])
```
